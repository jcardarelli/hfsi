```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Download NLTK data if necessary
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Preprocessing function
def preprocess_text(text):
    # Tokenize the text
    words = word_tokenize(text)
    
    # Initialize the lemmatizer
    lemmatizer = WordNetLemmatizer()
    
    # Remove punctuation and stopwords, and lemmatize
    stop_words = set(stopwords.words('english'))
    cleaned_words = [lemmatizer.lemmatize(word.lower()) for word in words if word.lower() not in stop_words and word not in string.punctuation]
    
    return ' '.join(cleaned_words)

# Example usage
processed_text = preprocess_text(oakland)
print(processed_text[:1000])
```

```python
import PyPDF2
import difflib
import os
from typing import List, Tuple
import argparse
from datetime import datetime

def extract_text_from_pdf(pdf_path: str) -> str:
    """
    Extract text content from a PDF file.
    
    Args:
        pdf_path (str): Path to the PDF file
        
    Returns:
        str: Extracted text content
    """
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ''
            for page in reader.pages:
                text += page.extract_text()
            return text
    except Exception as e:
        raise Exception(f"Error reading PDF {pdf_path}: {str(e)}")

def compare_pdfs(pdf_paths: List[str]) -> Tuple[List[str], str]:
    """
    Compare text content of three PDFs and generate differences.
    
    Args:
        pdf_paths (List[str]): List of paths to three PDF files
        
    Returns:
        Tuple[List[str], str]: List of differences and comparison report
    """
    if len(pdf_paths) != 3:
        raise ValueError("Exactly three PDF paths are required")
    
    # Extract text from all PDFs
    texts = []
    for path in pdf_paths:
        if not os.path.exists(path):
            raise FileNotFoundError(f"PDF file not found: {path}")
        texts.append(extract_text_from_pdf(path).splitlines())
    
    # Compare texts using difflib
    differences = []
    report = f"PDF Comparison Report - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n"
    
    # Compare each pair of PDFs
    pairs = [(0, 1), (1, 2), (0, 2)]
    for i, j in pairs:
        diff = list(difflib.unified_diff(
            texts[i], 
            texts[j],
            fromfile=os.path.basename(pdf_paths[i]),
            tofile=os.path.basename(pdf_paths[j]),
            lineterm=''
        ))
        
        if diff:
            differences.extend(diff)
            report += f"\nDifferences between {os.path.basename(pdf_paths[i])} and {os.path.basename(pdf_paths[j])}:\n"
            report += '\n'.join(diff) + '\n'
            report += '-' * 80 + '\n'
    
    if not differences:
        report += "\nNo differences found between the PDFs."
    
    return differences, report

def save_report(report: str, output_path: str):
    """
    Save the comparison report to a file.
    
    Args:
        report (str): Comparison report content
        output_path (str): Path to save the report
    """
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(report)

def main():
    parser = argparse.ArgumentParser(description='Compare three PDF files and find differences')
    parser.add_argument('pdf1', help='text-file/sanfran.pdf')
    parser.add_argument('pdf2', help='text-file/oakland.pdf')
#     parser.add_argument('pdf3', help='text-file/california.pdf')
    parser.add_argument('--output', '-o', default='pdf_comparison_report.txt',
                      help='Output file path for comparison report')
    
    args = parser.parse_args()
    
    try:
        differences, report = compare_pdfs([args.pdf1, args.pdf2, args.pdf3])
        save_report(report, args.output)
        print(f"\nComparison completed successfully. Report saved to: {args.output}")
        
        if differences:
            print("\nDifferences found between the PDFs. Check the report for details.")
        else:
            print("\nNo differences found between the PDFs.")
            
    except Exception as e:
        print(f"\nError: {str(e)}")
        exit(1)

if __name__ == "__main__":
    main()
```
