```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
from striprtf.striprtf import rtf_to_text

with open("text-files/oakland2.rtf", "r") as f:
    rtf_text = f.read()
    plain_text = rtf_to_text(rtf_text)

print(plain_text)
```

```python
# Example of basic preprocessing function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove non-alphabetic characters (you can keep punctuation if necessary)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Tokenize the text (optional, can be done later with vectorizers)
    words = text.split()
    
    return " ".join(words)

processed_text = preprocess_text(plain_text)
print(processed_text)
```

```python
# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the preprocessed text
X_tfidf = tfidf_vectorizer.fit_transform([processed_text])

# Convert the result to an array to see the vectorized form
vectorized_tfidf = X_tfidf.toarray()
print(vectorized_tfidf)

# Show the vocabulary (words with their TF-IDF scores)
print(tfidf_vectorizer.get_feature_names_out())
```

```python
df = pd.DataFrame(vectorized_tfidf, columns=tfidf_vectorizer.get_feature_names_out())
```
